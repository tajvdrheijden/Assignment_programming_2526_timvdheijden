## Model extension:
Based on the Solow model, the exogenous savings rule ($I_t = sY_t$) is replaced by an endogenous saving rule implied by the Euler consumption law (for more information about the Euler consumption law see [info](https://mitsloan.mit.edu/shared/ods/documents?DocumentID=4171)). Meaning that investment is not just a fixed fraction s (savings) of output, instead, consumption $c_t$ is chosen according to the household equation of Euler (where investment equals output minus consumption):
$$i_t = k_{t+1} - (1-\delta)k_t = f(k_t) - c_t$$
where consumption is based on the Euler equation:
$$\begin{align}
\frac{c_{t+1}}{c_t}&= [\beta(f^{\prime}(k_{t+1})+1-\delta)]^{1/\theta}
\end{align}$$

## the baseline Solow model:
The baseline Solow model implies that total output is as follows
$$Y_t = K^\alpha_tL^{1-\alpha}_t$$
and capital accumulation
$$K_{t+1} = sY_t + (1-\delta)K_t$$
with $L_t$ constant we derive a steady state that satisfies:
$$k_{t+1} = sAk^{*\alpha}/(\delta + n)$$
However that is when s is exogenous, therefore if we want to make s endogenous following the Euler equation we begin as follows:

**First** we assume that households maximizes lifetime utility by:
\begin{equation}
\max_{{c_t}} \sum_{t=0}^{\infty}\beta^t\frac{c^{1-\theta}_t - 1}{1-\theta} 
\end{equation}
Where $0<\beta<1$ and $\frac{c^{1-\theta}_t - 1}{1-\theta}$ can be written as $u(c_t)$

Which is subject to the capital accumulation constraint
$$k_{t+1} = f(k_t) + (1-\delta)k_t-c_t$$
where $f(k_t) = Ak^{\alpha}_t$

The **second** step is to perform the first-order condition of the Euler equation, by using the Langragian:
\begin{equation}
\mathcal{L} = \sum_{t=0}^{\infty}\beta^t[u(c_t) + \lambda(f(k_t) + (1-\delta)k_t - c_t - k_{t+1})]
\end{equation}
FOC w.r.t. $c_t$:
\begin{align}
\frac{\partial \mathcal{L}}{\partial c_t}: u^\prime(c_t)-\lambda_t = 0 \iff \lambda_t = u^\prime(c_t)
\end{align}
FOC w.r.t. $k_{t+1}$:
\begin{align}
\frac{\partial \mathcal{L}}{\partial k_{t+1}}: -\lambda_t+\beta\lambda_{t+1}(f^\prime(k_{t+1})+1-\delta)
\end{align}
substitute $\lambda_t = u^\prime(c_t)$ to get the Euler equation:
$$u^\prime(c_t)=\beta u^\prime(c_{t+1})(f^\prime(k_{t+1})+1-\delta)$$

**Thirdly**, combining the Euler equation with consumption growth rule (CRRA) results in,
$u^\prime(c)=c^{-\theta}$, meaning that the Euler equation becomes now:
$$c^{-\theta}_t=\beta c^{-\theta}_{t+1}(f^\prime(k_{t+1})+1-\delta) \iff (\frac{c_{t+1}}{c_t})^{\theta} = \beta(f^\prime(k_{t+1})+1-\delta) \iff \frac{c_{t+1}}{c_t} = [\beta(f^\prime(k_{t+1})+1-\delta)]^{1/\theta}$$

**Fourthly**, at the steady state we have $k_{t+1} = k_t = k^*$ and $c_{t+1} = c_t = c^*$, so we get:
$$u^{\prime}(c^*)=\beta u^\prime(c^*)(f^\prime(k^*)+1-\delta) \iff 1 = \beta(f^\prime(k^*)+1-\delta) \iff f^{\prime}(k^*)=\frac{1}{\beta}-1+\delta$$
Once we have obtained $k^*$, the steady state output is $y^*=f(k^*)$.
The consumption is then found from the resource constraint:
$$c^*=f(k^*)-\delta k^*$$
And the steady state $s^*$ is then:
$$s^*=\frac{investment^*}{y^*}= \frac{\delta k^*}{f(k^*)}$$

So we have now obtained endogenous savings function, as $k^*$ is determined by the Euler condition compared to previously an exogenous constant savings rate of $s$


## Code
First we need to define the parameters as above:

```python
# YOUR CODE HERE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

α = 0.33
β = 0.96
δ = 0.05
A = 1.0
θ = 2.0
T = 50
```

Then we define the production function ($f(k)$) and the marginal product function ($f^{\prime}(k)$):

```python
def f(k): return A * k**α
def fprime(k): return α * A * k**(α - 1)
```

As derived above by using the Euler equation we define the steady state $f^{\prime}(k^*)$ and solve for $k^*$ and $c^*$:

```python
rhs = 1.0 / β - 1.0 + δ
sol = root_scalar(lambda k: fprime(k) - rhs, bracket=[1e-8, 1e6], method='bisect')
k_star = sol.root
c_star = f(k_star) - δ * k_star

print("Steady state k* =", k_star)
print("Steady state c* =", c_star)
```

### Simulation to the steady state
If we choose to start below the steady state we use the following levels of $k_0$ and $c_0$

```python
k = np.zeros(T+1)
c = np.zeros(T+1)
k[0] = 0.89 * k_star
c[0] = f(k[0]) - δ * k[0] - 0.05
```

Simulate with the Euler consumption rule (endogenous savings):
$$c_{t+1} = c_t * \beta(f^\prime(k^*)+1-\delta)$$
While keeping in mind the corresponding resource constraint:
$$k_{t+1} = f(k_t) + (1-\delta)k_t - c_t$$

The resource constraint:

```python
for t in range(T):
    k[t+1] = f(k[t]) + (1 - δ) * k[t] - c[t]
    if k[t+1] <= 0:
        raise ValueError(f"Negative capital at t={t+1}, adjust value of k0 and/or co or parameters.")
```

Consumption growth by Euler:

```python
 c[t+1] = c[t] * (β * (fprime(k[t+1]) + 1 - δ))**(1.0/θ)
```

### Graphs
Lastly, we plot the results by having time on the x-axis and for capital (k) on the y-axis or for consumption (c) on the y-axis

```python
time = np.arange(T+1)
plt.figure(figsize=(10,6))
plt.plot(time, k, label='k_t (capital)')
plt.axhline(k_star, color='k', linestyle='--', label='k*')
plt.legend(); plt.title('Capital accumulation'); plt.xlabel('time'); plt.grid(True)
plt.show()

plt.figure(figsize=(10,6))
plt.plot(time, c, label='c_t (consumption)')
plt.axhline(c_star, color='k', linestyle='--', label='c*')
plt.legend(); plt.title('Consumption growth'); plt.xlabel('time'); plt.grid(True)
plt.show()

df = pd.DataFrame({'T': time, 'k': k, 'c': c})
df
```

We can also plot the implied saving rate, but first we have to calculate the steady state $s^*$ by the following formula: $$s^*=\frac{investment^*}{y^*}= \frac{\delta k^*}{f(k^*)}$$ 
Leading to the following code:

```python
investment = k[1:] - (1 - δ) * k[:-1]   
output = f(k[:-1])
saving_rate = investment / output
plt.figure(figsize=(10,5))
plt.plot(time[:-1], saving_rate, label='implied s_t')
plt.title('Implied saving rate (endogenous)')
plt.xlabel('time'); plt.grid(True); plt.legend()
plt.show()

ss = pd.DataFrame({'implied saving rate' : saving_rate})
ss
```

However, the economy does not converge to $k^*$ since the initial guess for consumption $c_0$ is not chosen on a specific path. As explained by the Ramsey-Cass-Koopmans (also known as Ramsey) model ([see here](https://web.econ.ku.dk/okocg/VV/VV-2010/Lecture%20notes/Ch7-2010-1.pdf)), the initial consumption $c_0$ has to be chosen to be inline on a specific path. If and only if $c_0$ lies on this path, an equilibrium of consumption can be found, resulting in the convergence of capital to its steady state. Yet, in this simulation, this is not the case, resulting in the overshooting of capital. Thereby, the marginal product of capital decreases, creating a fall in consumption and an explosion in savings (as shown in the graphs after $t=20$).

All in all, the extension model has shown that when using endogenous savings in the Solow model an optimal level for capital does theoretically exist.
